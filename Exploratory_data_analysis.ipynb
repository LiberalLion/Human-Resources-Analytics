```python
#Exploratory Data analysis
#Plotting all the attributes of the dataset against eachother

import pandas as pd
import numpy as np
from pandas import Series, DataFrame
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'C:\Users\***\HR_comma_sep.csv') #read data from csv
data.info() #data summary
train, test = train_test_split(data, test_size = 0.2)#split the training and testing set
```

```python
percentages_of_work_accident_leave = []
#calculate people's leaving and time spend's relationship.
stay_accident = people_stay[people_stay['Work_accident'] == 0]
left_accident = people_left[people_left['Work_accident'] == 0]
percentages_of_work_accident_leave.append(100*len(left_accident)*1.0/(len(left_accident) + len(stay_accident)))

stay__no_accident = people_stay[people_stay['Work_accident'] == 1]
left_no_accident = people_left[people_left['Work_accident'] == 1]
percentages_of_work_accident_leave.append(100*len(left_no_accident)*1.0/(len(left_no_accident) + len(stay__no_accident)))

accident = [0,1]

plt.rc('xtick', labelsize=20)    # fontsize of the tick labels
plt.rc('ytick', labelsize=30)    # fontsize of the tick labels
plt.figure(figsize=(20,10))
plt.bar(accident, percentages_of_work_accident_leave)
plt.title("working accident VS. the percentages of leaving",fontsize = 50)

plt.show()
```

```python
#plot the sales type and leaving
pecentages_total = train.groupby(['promotion_last_5years'])['left'].count().reset_index()
pecentages_left = train.groupby(['promotion_last_5years'])['left'].agg(lambda x: (x==1).sum()).reset_index()
Proportion_of_merge = pd.merge(pecentages_left,pecentages_total ,on = 'promotion_last_5years')
Proportion_of_merge["Proportion"] = (100.0*Proportion_of_merge['left_x'] / Proportion_of_merge['left_y'])
plt.rc('xtick', labelsize=20)    # fontsize of the tick labels
plt.rc('ytick', labelsize=30)    # fontsize of the tick labels
plt.figure(figsize=(20, 10))
sns.barplot(x='promotion_last_5years',y='Proportion',data=Proportion_of_merge)
plt.xticks(rotation=90)
plt.title("promotion_last_5years vs leaving percentages",fontsize = 50)

plt.show()
```
